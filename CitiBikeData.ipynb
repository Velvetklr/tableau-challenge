```python
import pandas as pd
import csv
import numpy as np
import datetime as dt
```

```python
jan2022 = "Resources/JC-202201-citibike-tripdata.csv"
mar2022 = "Resources/JC-202203-citibike-tripdata.csv"
aug2022 = "Resources/JC-202208-citibike-tripdata.csv"
oct2022 = "Resources/JC-202210-citibike-tripdata.csv"

jan2023 = "Resources/JC-202301-citibike-tripdata.csv"
mar2023 = "Resources/JC-202303-citibike-tripdata.csv"
aug2023 = "Resources/JC-202308-citibike-tripdata.csv"
oct2023 = "Resources/JC-202310-citibike-tripdata.csv"

```

```python
jan_2022 = pd.read_csv(jan2022)
mar_2022 = pd.read_csv(mar2022)
aug_2022 = pd.read_csv(aug2022)
oct_2022 = pd.read_csv(oct2022)

jan_2023 = pd.read_csv(jan2023)
mar_2023 = pd.read_csv(mar2023)
aug_2023 = pd.read_csv(aug2023)
oct_2023 = pd.read_csv(oct2023)
```

```python
jan_2022.head()
```

```python
citi_concat = pd.concat([jan_2022, mar_2022, aug_2022, oct_2022,
                         jan_2023, mar_2023, aug_2023, oct_2023], ignore_index=True)
citi_concat
```

```python
citi_concat.count()
```

```python
drop_citi_bike = citi_concat.dropna()
```

```python
def parse_time(raw):
    date = dt.datetime.fromisoformat(raw).date()
    string = date.isoformat()
    return string
```

```python
# drop_citi_bike.loc[:,'start_date'] = drop_citi_bike['started_at'].map(parse_time)
# drop_citi_bike
```

```python
drop_citi_bike.loc[:,'started_at'] = pd.to_datetime(drop_citi_bike['started_at'],infer_datetime_format=True)
```

```python
drop_citi_bike.loc[:,'ended_at'] = pd.to_datetime(drop_citi_bike['ended_at'],infer_datetime_format=True)
```

```python
drop_citi_bike.loc[:,'trip_duration'] = drop_citi_bike['ended_at']- drop_citi_bike['started_at']
```

```python
drop_citi_bike
```

```python
def delta(raw):
    convert = raw.seconds
    return convert

```

```python
def deltaday(raw):
    convert = raw.days
    return convert
```

```python
drop_citi_bike.loc[:,'trip_duration(secs)'] = drop_citi_bike['trip_duration'].map(delta)
drop_citi_bike.loc[:,'trip_duration(day)'] = drop_citi_bike['trip_duration'].map(deltaday)
drop_citi_bike
```

```python
sorted_df = drop_citi_bike.sort_values('trip_duration(secs)',ascending=False)
sorted_df
```

```python
drop = sorted_df['trip_duration(day)']== -1
```

```python
sorted_df = sorted_df[~drop]
sorted_df
```

```python
sorted_df.head(30)

sorted_df = sorted_df.drop(['trip_duration(day)','trip_duration'], axis=1)
sorted_df
```

```python
clean_citi_bike = sorted_df.reset_index()
```

```python
clean_citi_bike.to_csv("clean_citi_bike.csv",index=False)
```
